```python
# make a dataframe
# col0: time, 0-1s, 1000 steps
# col1: sin, sine wave of time, 10 Hz
# col2: cos, cosine wave of time, 10 Hz
# col3: log, log of time
import numpy as np
import pandas as pd
df = pd.DataFrame()
df['time'] = np.linspace(0, 1, 1000)
df['sin'] = np.sin(2*np.pi*10*df['time'])
df['cos'] = np.cos(2*np.pi*10*df['time'])
df['log'] = np.log(df['time'])
print(df)
```

```python
# make plot of above dataframe
import matplotlib.pyplot as plt
plt.plot(df['time'], df['sin'], label='sin')
plt.plot(df['time'], df['cos'], label='cos')
plt.plot(df['time'], df['log'], label='log')
plt.legend()
plt.show()
```
